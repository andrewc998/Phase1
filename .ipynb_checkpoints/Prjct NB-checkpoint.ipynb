```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.cm as cm 
```

```python
#loading dataset and showing first 5
df=pd.read_csv("AviationData.csv",encoding="latin1",low_memory=False)
#display first 5
df.head()
```

```python
df.columns #shows the columns in the dataset
```

```python
df.info() #helps us check the non null values
```

```python
# Summary statistics of numeric columns
df.describe()
```

```python
#Convert date columns to datetime format
df['Event.Date'] = pd.to_datetime(df['Event.Date'], errors='coerce')
```

```python
# Checking the number of missing values in each column
df.isna().sum()
```

```python
df.isnull().sum() / len(df) * 100 #checks percentage of missing columns to know what data to remove
```

```python
#dropping data that may not be useful for our objectives and those with high missing values - well over 50pc

df.drop(['Event.Id','Accident.Number','Airport.Code','Latitude','Longitude',
         'Airport.Name','Location','Registration.Number',
         'FAR.Description','Report.Status','Schedule','Air.carrier','Publication.Date'],axis=1,inplace=True)
df.info()
```

```python
df.isnull().sum() / len(df) * 100 #checks percentage of missing columns
```

```python
most_common_category = df['Aircraft.Category'].mode()[0] 
df['Aircraft.Category'] = df['Aircraft.Category'].fillna(most_common_category)
```

```python
most_common_phase = df['Broad.phase.of.flight'].mode()[0]
df['Broad.phase.of.flight'] = df['Broad.phase.of.flight'].fillna(most_common_phase)
```

```python
# List of columns to fill
columns_to_fill = ['Injury.Severity', 'Aircraft.damage', 'Engine.Type', 'Weather.Condition']

# Iterate through each column
for column in columns_to_fill:
    # Find the most frequent value
    most_common_value = df[column].mode()[0]
    
    # Fill missing values
    df[column] = df[column].fillna(most_common_value)
    
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
print(categorical_cols)
for col in categorical_cols:
    num_missing = df[col].isnull().sum()
    if num_missing > 0:
        print(f"{col}: {num_missing} missing values")
```

```python
# I choose to fill in remaining values with 'Unknown' as using the median may skew data 
# on best risk policies to make

# List of columns to fill with 'Unknown'
columns_to_fill_unknown = ['Country', 'Make', 'Model', 'Amateur.Built', 'Purpose.of.flight']

# Fill missing values with 'Unknown'
for column in columns_to_fill_unknown:
    df[column] = df[column].fillna('Unknown')
```

```python
# Select numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns

# Check for missing values in numerical columns
for col in numerical_cols:
    num_missing = df[col].isnull().sum()
    if num_missing > 0:
        print(f"{col}: {num_missing} missing values")
```

```python
# Chose to use median to fill missing values to avoid 
#outliers, for data integrity and data skewing(abnormal distribution)

# List of numerical columns with missing values
numerical_cols_with_missing = ['Total.Fatal.Injuries', 'Total.Serious.Injuries', 
                               'Total.Minor.Injuries','Total.Uninjured','Number.of.Engines']  
# Replace with your list

# Fill missing values with the median
for col in numerical_cols_with_missing:
    median_value = df[col].median()
    df[col] = df[col].fillna(median_value)
```

```python

#add an 'Accident.Number' column to  uniquely identify each accident
df['Accident.Number'] = df.index + 1
```

```python
#Adding total injuries column 
df['Total.Injuries'] = df[['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries']].sum(axis=1)
df.info()
```

```python
#The Risk Score is a way to quantify the overall risk of incidents or accidents based on the severity of injuries.

#Total.Fatal.Injuries: Each fatal injury contributes 3 points to the risk score.

#Total.Serious.Injuries: Each serious injury contributes 2 points.

#Total.Minor.Injuries: Each minor injury is 1point
```

```python
# Generate Risk Score column
df["Risk.Score"] = (
    df["Total.Fatal.Injuries"] * 3 +
    df["Total.Serious.Injuries"] * 2 +
    df["Total.Minor.Injuries"] * 1
)
```

```python
# Compute Accident Rate per Aircraft Model
#Accident Rate= accidents per model / total records per model
#Identifies aircraft with recurrent issues

accident_counts = df.groupby(["Make", "Model"]).size().reset_index(name="Accident.Count")
total_records = df.groupby(["Make", "Model"])["Accident.Number"].count().reset_index(name="Total.Records")
accident_rate = pd.merge(accident_counts, total_records, on=["Make", "Model"])
accident_rate["Accident.Rate"] = accident_rate["Accident.Count"] / accident_rate["Total.Records"]

# Merge Accident Rate back to the main dataset
df = df.merge(accident_rate[["Make", "Model", "Accident.Rate"]], on=["Make", "Model"], how="left")
```

```python
df.rename(columns={'Accident.Rate': 'Accident.Rate.Per.Craft.Model'}, inplace=True)
df.info()
```

```python
#Flight Purpose Risk Score
# ----Some purposes (e.g., training, personal flights) might be riskier.
#Compute the average Risk Score per flight purpose to identify high-risk activities.


# Calculate the average Risk Score for each type of flight purpose
risk_score_per_flight_pupose = df.groupby("Purpose.of.flight")["Risk.Score"].mean().reset_index()
risk_score_per_flight_pupose.rename(columns={"Risk.Score": "Flight.Purpose.Risk"}, inplace=True)

# Merge the computed Flight Purpose Risk Score back into the main dataset
df = df.merge(risk_score_per_flight_pupose, on="Purpose.of.flight", how="left")
df.info()
```

```python
#Survivability Rate (Percentage of passengers uninjured per accident)
# -------Helps show crafts with better safety overall
df["Survivability.Rate.Pc"] = (df["Total.Uninjured"] / (df["Total.Injuries"] + df["Total.Uninjured"])) * 100
```

```python
#Fatality Rate per Aircraft (Percentage of injuries that resulted in fatalities)
# ------------ Can help identify models with low fatality risk
df["Fatality.Rate.Per.Aircraft.Pc"] = (df["Total.Fatal.Injuries"] / df["Total.Injuries"]) * 100
```

```python
#Compute Phase of flight Risk Score
# ---------------------------------
# Assign risk weights to flight phases
phase_risk_weights = {
    "Takeoff": 3,
    "Approach": 3,
    "Landing": 3,
    "Cruise": 2,
    "Maneuvering": 2,
    "Taxi": 1,
    "Standing": 1
}

# Map risk scores based on phase of flight
df["Flight.Phase.Risk.Score"] = df["Broad.phase.of.flight"].map(phase_risk_weights)
```

```python
# Compute Weather Risk Factor
# -------------------------------------
weather_risk_scores = {
    "IMC": 5,  # Highest risk
    "Marginal VMC": 3,  # Moderate risk
    "VMC": 1  # Low risk
}

# Map numerical risk scores for weather conditions
df["Weather.Risk.Factor"] = df["Weather.Condition"].map(weather_risk_scores)
```

```python
# Fill missing values of new columns with appropriate strategies
df['Survivability.Rate.Pc'] = df['Survivability.Rate.Pc'].fillna(df['Survivability.Rate.Pc'].median())
df['Flight.Phase.Risk.Score'] = df['Flight.Phase.Risk.Score'].fillna(df['Flight.Phase.Risk.Score'].median())
df['Weather.Risk.Factor'] = df['Weather.Risk.Factor'].fillna(df['Weather.Risk.Factor'].median())



df.isnull().sum().sum()  # Should return 0 if all missing values are addressed
```

Shows the industry might be getting safer for new businesses.

```python
# try to find the safest engine based on numbers


# Filter out invalid engine types
valid_engine_types = ["Reciprocating", "Turbo Fan", "Turbo Shaft", "Turbo Prop", "Turbo Jet",
                      "Electric", "Hybrid Rocket", "Geared Turbofan"]

# Group by Engine Type and calculate safety metrics
engine_safety = df[df["Engine.Type"].isin(valid_engine_types)].groupby("Engine.Type")[
    ["Survivability.Rate.Pc", "Fatality.Rate.Per.Aircraft.Pc", "Flight.Purpose.Risk"]
].mean().reset_index()

# Sort by safety (highest survivability, lowest fatality, lowest risk)
safest_engines = engine_safety.sort_values(
    by=["Survivability.Rate.Pc", "Fatality.Rate.Per.Aircraft.Pc", "Flight.Purpose.Risk"],
    ascending=[False, True, True]
)

safest_engines
```

```python
# Define airplane engine types for Commercial and Private flights with less risks
#first group commercial and private categories
commercial_categories = ["Executive/corporate", "Business", "Public Aircraft", "Ferry",
                         "Public Aircraft - Federal", "Public Aircraft - Local", "Public Aircraft - State"]
private_categories = ["Personal", "Instructional"]

# Group by Aircraft Make and calculate mean survivability rate and risk metrics
aircraft_safety_by_make = df.groupby('Make').agg({
    'Survivability.Rate.Pc': 'mean',
    'Risk.Score': 'mean',
    'Flight.Purpose.Risk': 'mean',
    'Engine.Type': lambda x: x.mode()[0] if not x.mode().empty else "Unknown"  # Most common engine type
}).reset_index()

# Identify the safest aircraft makes for commercial flights
safest_commercial_makes = aircraft_safety_by_make[df["Purpose.of.flight"].isin(commercial_categories)].sort_values(
    by=["Survivability.Rate.Pc", "Flight.Purpose.Risk", "Risk.Score"],
    ascending=[False, True, True]
).head(10)

# Identify the safest aircraft makes for private flights
safest_private_makes = aircraft_safety_by_make[df["Purpose.of.flight"].isin(private_categories)].sort_values(
    by=["Survivability.Rate.Pc", "Flight.Purpose.Risk", "Risk.Score"],
    ascending=[False, True, True]
).head(10)

safest_commercial_makes [['Make', 'Flight.Purpose.Risk', 'Engine.Type']]
safest_private_makes[['Make', 'Flight.Purpose.Risk', 'Engine.Type']]
```

```python
# Further try and Filter turbo fan aircraft with commercial purposes
turbo_commercial = df[(df['Engine.Type'] == 'Turbo Fan') &
                      (df['Purpose.of.flight'].isin(['Executive/corporate', 'Unknown']))]  # Assuming 'Unknown' may include some commercial ops

# Calculate fatal accident rate per model
turbo_stats = turbo_commercial.groupby(['Make', 'Model']).agg({
    'Total.Fatal.Injuries': 'sum',
    'Total.Uninjured': 'sum',
    'Accident.Number': 'count'
}).reset_index()

# Calculate metrics
turbo_stats['Fatal_Accident_Rate'] = turbo_stats['Total.Fatal.Injuries'] / turbo_stats['Accident.Number']
turbo_stats['Survival_Rate'] = turbo_stats['Total.Uninjured'] / (turbo_stats['Total.Uninjured'] + turbo_stats['Total.Fatal.Injuries'])

# Plot
plt.figure(figsize=(12, 6))
sns.scatterplot(data=turbo_stats,
                x='Fatal_Accident_Rate',
                y='Survival_Rate',
                size='Accident.Number',
                hue='Make',
                sizes=(100, 500),
                alpha=0.8,
                palette='viridis')
plt.axvline(x=turbo_stats['Fatal_Accident_Rate'].median(), linestyle='--', color='gray')
plt.axhline(y=turbo_stats['Survival_Rate'].median(), linestyle='--', color='gray')
plt.title('Turbo Fan Aircraft Safety Profile (Commercial Operations)', fontsize=14)
plt.xlabel('Fatal Accidents per Incident')
plt.ylabel('Survival Rate (%)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show();
```

```python
#Lower right quadrant: High fatality rate, low survival (avoid)
#Upper left quadrant: Low fatality rate, high survival (target)
#Bubble size represents sample size (more reliable results)
```

```python
# Try to break down the best commercial airplanes based on some metrics (survivability and fatality)
candidates = ['Boeing 737', 'Boeing 727', 'DC-10', 'FALCON 20']
metrics = turbo_commercial[turbo_commercial['Model'].isin(candidates)].groupby('Model').agg({
    'Total.Fatal.Injuries': 'sum',
    'Total.Uninjured': 'sum',
    'Accident.Number': 'count'
}).sort_values('Total.Fatal.Injuries')

# Calculate rates
metrics['Fatality_Rate'] = metrics['Total.Fatal.Injuries'] / metrics['Accident.Number']
metrics['Survival_Rate'] = metrics['Total.Uninjured'] / (metrics['Total.Uninjured'] + metrics['Total.Fatal.Injuries'])

# Plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Fatality Rate
sns.barplot(y=metrics.index, x='Fatality_Rate', data=metrics,
            palette='Reds_r', ax=ax1)
ax1.set_title('Fatality Rate per Incident')
ax1.set_xlabel('Fatalities per Accident')
ax1.axvline(x=metrics['Fatality_Rate'].median(), color='black', linestyle='--')

# Survival Rate
sns.barplot(y=metrics.index, x='Survival_Rate', data=metrics,
            palette='Greens_r', ax=ax2)
ax2.set_title('Survival Rate')
ax2.set_xlabel('Percentage Survived')
ax2.axvline(x=metrics['Survival_Rate'].median(), color='gray', linestyle='--')

plt.suptitle('Turbo Fan Fleet Safety Comparison', fontsize=16, y=1.02)
plt.tight_layout()
plt.show();
```
